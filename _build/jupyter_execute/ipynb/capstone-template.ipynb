# Projeto Final de Visualização de Dados - PFVD

Siga este documento orientador para saber como organizar-se para a avaliação do Módulo 3 do curso.

## Orientações Gerais

### Definição 

O Projeto Final de Visualização de Dados (PFVD) é uma tarefa desempenhada de forma individual ou grupal (no máximo 4 integrantes) que aplica todo o conhecimento adquirido no curso para a entrega de dois ativos: i) relatório formal e ii) produto de visualização material.

### Objetivo 

Empregar o conhecimento adquirido para elaborar um produto de visualização útil a uma audiência pré-selecionada cuja materialização dá-se, mas não se limita, por:
- relatórios ou laudos automatizados; ou
- projetos visuais aplicados ou _dashboards_; ou
- ideação de gráficos ou elementos visuais inovadores.

### Escopo

O PFVD deverá se sustentar pela base de conhecimento adquirida a partir dos Módulos 1 a 3 do curso, bem como pela disponível em fontes externas.

### Tema

Livre.

### Ativos

#### Relatório formal

O documento formal (relatório) que sustentará a base de conhecimento teórico-metodológica do PFVD deverá ser redigido em até 4 páginas, contendo as seguintes seções:

 - _Escopo_: campo de conhecimento científico ou domínio de expertise no qual o projeto se insere;
 - _Desafio_: descrição do desafio (científico, tecnológico, operacional etc.) que o projeto busca solucionar;
 - _Projeto de visualização_: descrição das etapas a serem tomadas para atingimento da representação visual (documento descritivo do modelo referencial);
 - _Mapeamento_: codificação resumida utilizada para o mapeamento dos dados de entrada;
 - _Representação(ões) visual(is)_ (RV(s)): plotagem(ns) obtida(s) após a execução do previsto no projeto de visualização;
 - _Análise de dados_: discussão resumida da leitura analítica que pode ser feita a partir da(s) RV(s);
 - _Takeaways_: síntese do mini-projeto em três pontos de destaque principais ("mensagem a ser levada para casa");
 - _Caracterização do produto mínimo viável_: síntese das funções práticas do produto de visualização proposto.

#### Produto de visualização material

- O produto material e suas funcionalidades será demonstrado por meio de apresentação (em até 10 minutos) para a classe.
 
### Nota

No PFVD, a ênfase está na **entrega de um produto de visualização** que demonstre de maneira clara uma oferta de solução para o desafio proposto.


# Estrutura proposta para o relatório

## Identificação

**Título:** 

- Título do relatório no formato "PFVD: NOME DA APLICAÇÃO / PRODUTO / SOLUÇÃO. Exemplo:
    - _PFVD: DATA REPORTING APLICADO AO DIAGNÓSTICO DE CÂNCER DE MAMA_

**Equipe:** 

- Nomes e emails de contato dos membros da equipe. Exemplo:
    - Euclides de Alexandria, 'euclides.alexandria@nome.edu.br'
    - Rei Henrique VII, 'henry@eu.com'
    - Cosme de Médici, 'cmedici@yuhh.com.br'

## Resumo

Escreva em uma ou duas sentenças, no máximo, a finalidade do relatório.

## Conteúdo do Relatório 

### Escopo

Defina o campo ou domínio de conhecimento que o produto visa alcançar.

### Desafio

Descreva a situação-problema ou a lacuna que o produto visa resolver, corrigir ou aperfeiçoar.

### Projeto de visualização

Ilustre o fluxo de processos que foram seguidos para a geração do produto. 

### Mapeamento

Explique como o produto fará a associação entre os dados fundamentais de que ele depende para funcionar e os elementos visuais que produzirá.

### Representações visuais

Descreva que tipo(s) de RV(s) o produto abordará e exemplifique a aplicabilidade delas.

### Análise de dados

Discuta o que a(s) RV(s) produzidas pelo produto mostram quando aplicadas a um caso particular, interpretando-a(s) minimamente.

### _Takeaways_

Sintetize em até três pontos de destaque principais ("mensagem a ser levada para casa") porque o produto de visualização é relevante.

### _Caracterização do produto mínimo viável_

Especifique resumidamente as ferramentas tecnológicas utilizadas para a elaboração do produto, plataforma, metadados, modo de utilização, repositório de hospedagem de código ou demais informações relevantes, no que couber.

### Referências bibliográficas

Referencie todas as fontes utilizadas que não forem de sua autoria, as quais podem incluir, mas não se limitam a:

- Aplicações e software de terceiros;
- Artigos técnicos e opiniões de experts;
- Blogs e posts;
- etc.
